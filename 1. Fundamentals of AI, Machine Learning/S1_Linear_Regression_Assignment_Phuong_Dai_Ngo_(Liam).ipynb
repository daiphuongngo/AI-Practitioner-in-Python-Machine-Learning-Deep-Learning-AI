```
# Method 1: Python
import sys
import numpy
import scipy
from matplotlib_venn import venn3,venn3_circles
from matplotlib import pyplot as plt
import seaborn as sns
sns.set() # create framework in white

#Build three lists to make 3 way venn diagram with                                                                                                                             
set1 = set(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r'])
set2 = set(['e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm'])
set3 = set(['h', 'i', 'j', 'k'])

#Make venn diagram                                                                                                                                                             
vd = venn3([set1,set2,set3],set_labels=("Artificial Intelligence","Machine Learning","Deep Learning"))

# Change font size of labels
for t in vd.set_labels: t.set_fontsize(15)

#Colours: get the HTML codes from the net  
# Palette: https://www.w3schools.com/colors/colors_shades.asp
# Position id: https://towardsdatascience.com/how-to-create-and-customize-venn-diagrams-in-python-263555527305
vd.get_patch_by_id("100").set_color("#3399FF")
vd.get_patch_by_id("110").set_color("#33CCFF")
vd.get_patch_by_id("111").set_color("#99FFFF")

# vd.get_patch_by_id("100").set_color("#c33c7d")
# vd.get_patch_by_id("110").set_color("#d576a4")
# vd.get_patch_by_id("111").set_color("#e8b5ce")

# vd.get_patch_by_id("100").set_color("#FF8000")
# vd.get_patch_by_id("110").set_color("#5858FA")
# vd.get_patch_by_id("111").set_color("#01DF3A")

# vd.get_patch_by_id("100").set_color("#b5cfe8")
# vd.get_patch_by_id("110").set_color("#b5dbe8")
# vd.get_patch_by_id("111").set_color("#b5e8e8")

#Move the numbers in the circles (no need)                                                                                                                                              
# vd.get_label_by_id("100").set_x(-0.1)
# vd.get_label_by_id("110").set_x(0.1)

#Strength of color, 2.0 is very strong.                                                                                                                                        
vd.get_patch_by_id("100").set_alpha(10)
vd.get_patch_by_id("110").set_alpha(10)
vd.get_patch_by_id("111").set_alpha(10)

# Remove data by set_test = ''
# Source: https://stackoverflow.com/questions/55622820/python-venn-diagram-actual-label
vd.get_label_by_id('100').set_text('')
vd.get_label_by_id('110').set_text('')
vd.get_label_by_id('111').set_text('')

plt.title("Relation ship between AI, ML and DL",fontsize=15)
plt.savefig("output",format="pdf")

# Relocate postion of labels # A, B, C as AI, ML, DL
lbl = vd.get_label_by_id("A")
x, y = lbl.get_position()
lbl.set_position((x+0.8, y-0.05))

lbl = vd.get_label_by_id("B")
x, y = lbl.get_position()
lbl.set_position((x-0.5, y-0.1))

lbl = vd.get_label_by_id("C")
x, y = lbl.get_position()
lbl.set_position((x+0.2, y+0.35))

# Referece:
#sns.palplot(sns.light_palette("navy", reverse=True))
```

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error as MSE
import matplotlib.pyplot as plt
np.random.seed(1612)
x = np.random.randn(2000,1)
noise = np.random.randn(2000,1)
y = 5* x**3 + 3 * x**2 + 10*noise + 2
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,shuffle=True,random_state=1612)
model = LinearRegression() 
model.fit(x_train, y_train) 
y_pred = model.predict(x_test) # !!! x_test, not x_train into model to has the same shape when plotting
plt.scatter(x_test, y_test, s=10)
plt.plot(x_test, y_pred, color='r')
plt.show()
```

```
print('Mean Absolute Error of Test Set:', MAE(y_test,y_pred))
```

```
print('Mean Squared Error of Test Set:', MSE(y_test, y_pred))
```
