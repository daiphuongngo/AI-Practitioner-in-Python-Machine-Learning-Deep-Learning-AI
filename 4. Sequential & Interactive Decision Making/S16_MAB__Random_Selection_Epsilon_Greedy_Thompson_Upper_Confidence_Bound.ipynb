## 16.1.1 [1 Point] MABs objective function
- What do multi-armed bandit algorithms try to maximize? Why is it a special case of MDP learning?
- Bài toán MAB tối ưu cái gì? Vì sao nó là trường hợp đặc biệt của bài toán học-ra-quyết-định trong MDP?

In probability theory, the multi-armed bandit problem (sometimes called the K- or N-armed bandit problem) is a problem in which a fixed limited set of resources must be allocated between competing (alternative) choices in a way that maximizes their expected gain, when each choice's properties are only partially known at the time of allocation, and may become better understood as time passes or by allocating resources to the choice. This is a classic reinforcement learning problem that exemplifies the exploration–exploitation tradeoff dilemma. The multi-armed bandit problem also falls into the broad category of stochastic scheduling.

## 16.1.2 [1 Point] MAB algorithms
- What are the similarity and differences between MAB algorithms? Describe specifically for the learned algorithms: $\epsilon$-greedy, Thompson sampling, UCB. 
- Liệt kê các điểm giống & khác giữa các giải thuật MAB, cụ thể cho 3 giải thuật đã được học. 

Three of the most popular MAB algorithms are Epsilon Greedy, Thompson Sampling, and Upper Confidence Bound 1 (UCB-1), each favoring exploitation over exploration to different degrees. 

Epsilon-Greedy

This is an algorithm for continuously balancing exploration with exploitation. (In ‘greedy’ experiments, the lever with highest known payout is always pulled except when a random action is taken). A randomly chosen arm is pulled a fraction ε of the time. The other 1-ε of the time, the arm with highest known payout is pulled. Epsilon-Greedy is the greediest algorithm and has the highest regret over multiple trails among the three. 

Thompson Sampling (Bayesian)

With this randomized probability matching strategy, the number of pulls for a given lever should match its actual probability of being the optimal lever. This is the best algorithm having the most stable regret line over trails and the highest reward afterall.

Upper Confidence Bound

This strategy is based on the Optimism in the Face of Uncertainty principle, and assumes that the unknown mean payoffs of each arm will be as high as possible, based on observable data.

```
def do_random_selection():
  # TODO: mô phỏng trường hợp khách hàng click vào quảng cáo với giá trị CTR thực
  return np.random.randint(0, 3)

class Stat:
    def __init__(self):
        # TODO: tham khảo code tại lớp và viết lại cho trường hợp có 3 quảng cáo
        self.index_list = []
        self.impressions = [0,0,0]
        self.clicks = [0,0,0]
        self.regret = 0 
        self.total_reward = 0
        ## danh sách ghi lại các giá trị regret trong mỗi lượt chiếu
        self.regret_list = []
        ## lists for collecting the calculated CTR 
        self.ctr = {0: [], 1: [], 2: []}
  
    def record(self, ad_index, did_click, ctrs):
        # TODO: tham khảo code tại lớp và viết lại cho trường hợp có 3 quảng cáo
        self.index_list.append(ad_index) # thông số báo ngoài nên ko cần bỏ self vào
        
        # TODO: Thống kê lại số lượt chiếu của mỗi quảng cáo
        # code here
        self.impressions[ad_index] += 1
        
        # TODO: Thống kê lại số lần khách hàng click vào mỗi quảng cáo
        if did_click: # True or False
            # code here
            self.clicks[ad_index] += did_click # True: += 1, False: += 0

        if ctrs and len(ctrs) > 0: # ở ngoài nạp vào CTR 
            self.ctr[0].append(ctrs[0])
            self.ctr[1].append(ctrs[1])
            self.ctr[2].append(ctrs[2])
        else: # tự tính CTR # do truyền ctrs = None thì ko chạy if ở trên mà chạy else này
            ## tính giá trị CTR và thêm nó vào danh sách
            if self.impressions[0] == 0:
                ctr_0 = 0
            else:
                ctr_0 = self.clicks[0]/self.impressions[0]
            if self.impressions[1] == 0:
                ctr_1 = 0
            else:
                ctr_1 = self.clicks[1]/self.impressions[1]    
            if self.impressions[2] == 0:
                ctr_2 = 0
            else:
                ctr_2 = self.clicks[2]/self.impressions[2]   
            self.ctr[0].append(ctr_0)
            self.ctr[1].append(ctr_1)
            self.ctr[2].append(ctr_2)
        
        ## tính toán regret và reward
        self.regret += max(ACTUAL_CTR) - ACTUAL_CTR[ad_index]
        self.regret_list.append(self.regret)
        self.total_reward += did_click
        
    def export_stat_dict(self):
        return {
            'reward': self.total_reward, 
            'regret_list': self.regret_list, 
            'ads_count': pd.Series(self.index_list).value_counts(normalize=True)
        }
```

```
# prob: the probability that customer will click on the ad
# return true/false
def simulate_customer(prob):
    # TODO: chọn ngẫu nhiên giá trị nguyên trong khoảng [0,1]
    return bernoulli.rvs(prob)

stat1 = Stat()
for i in range(n):    
    ad_index = do_random_selection()
    did_click = simulate_customer(ACTUAL_CTR[ad_index])                
    stat1.record(ad_index, did_click, None)
```

```
def do_epsilon_greedy(win_index, lose_indice, epsilon):
    # TODO: Bước 1: Lấy mẫu để quyết định xem lượt này là sẽ là khai phá (explore) hay khai thác (exploit)
    # 1: explore
    explore_mode = bernoulli.rvs(epsilon) # Bernoulli returns 0 or 1
    if explore_mode:
        # TODO: trả ra quảng cáo ít được chiếu hơn
        return random.choice(lose_indice) #np.argmin(lose_indice)
    # TODO: trả ra quảng cáo đang có ctr cao hơn
    return win_index # False: 0 

stat2 = Stat()
for i in range(n):
    lose_indice = [0,1,2]
    lose_indice.remove(win_ad_index)

    ad_index = do_epsilon_greedy(win_ad_index, lose_indice, e) # lose_indice or 1-win_ad_index ?
    did_click = simulate_customer(ACTUAL_CTR[ad_index])

    stat2.record(ad_index, did_click, None)
```

```
def do_thompson(priors, stat):
    if len(stat.index_list) == 0:
    # TODO: tham khảo code tại lớp và viết lại cho trường hợp có 3 quảng cáo
        return np.random.randint(0,3,1)[0], (0, 0, 0)

    # TODO: Lấy mẫu ctr_0, và ctr_1 dựa trên công thức:
    # 𝛼 = prior + số lượt click
    # 𝛽 = prior + số bỏ lỡ (tức là impressions - clicks)
    ctr_0 = random.betavariate(priors[0]+stat.clicks[0], priors[1] + stat.impressions[0] - stat.clicks[0])
    ctr_1 = random.betavariate(priors[1]+stat.clicks[1], priors[1] + stat.impressions[1] - stat.clicks[1])
    ctr_2 = random.betavariate(priors[2]+stat.clicks[2], priors[2] + stat.impressions[2] - stat.clicks[2])

    # TODO: Chọn quảng cáo có ctr lấy mẫu cao hơn
    win_index = np.argmax([ctr_0, ctr_1, ctr_2]) # ctr_2 #lấy CTR lớn nhất của lần lấy mẫu
    return win_index, (ctr_0, ctr_1, ctr_2) # ctr_2

stat3 = Stat()
for i in range(n):    
    win_index, ctrs = do_thompson(priors, stat3)
    did_click = simulate_customer(ACTUAL_CTR[win_index])
    stat3.record(win_index, did_click, ctrs)
```
